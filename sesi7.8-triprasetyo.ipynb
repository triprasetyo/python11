# Use Case 3 : Compare top 5 countries that has most immigrations to Canada

```python
df_can.sort_values('Total',ascending=False,inplace=True)

df_top5 = df_can.head(5)

df_top5 = df_top5[years].T

df_top5.index = df_top5.index.astype(int)

df_top5.plot(kind='line',figsize=(15,10))

plt.title('Top 5 Immigrant Countries')
plt.ylabel('No. of Immigrants') #Judul sumbu y
plt.xlabel('Years') #Judul sumbu x
plt.show()
```

# WordClouds

```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
```

```python
alice_novel = open('alice_novel.txt','r').read()
```

```python
stopwords = set(STOPWORDS)
```

```python
stopwords
```

```python
alice_wc = WordCloud(background_color='white',
                     max_words=2000,
                     stopwords=stopwords)

# generate world cloud
alice_wc.generate(alice_novel)
```

```python
fig = plt.figure(figsize=(10,10))
plt.imshow(alice_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
stopwords.add('said')
alice_wc = WordCloud(background_color='white',
                    max_words=2000,
                    stopwords=stopwords)

#generate word cloud
alice_wc.generate(alice_novel)
fig = plt.figure(figsize=(10,10))
plt.imshow(alice_wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```

# Regresion Plot

```python
!pip install seaborn
```

```python
import seaborn as sns
```

```python
df_total = df_can[years].sum(axis=0).to_frame()
df_total.index = df_total.index.astype(float)
df_total.reset_index(inplace=True)
df_total.columns = ['year','total']
df_total
```

```python
ax = sns.regplot(x='year',y='total',data=df_total)
ax.set_title('Regression Plot total Immigrants 1980-2010')
```

```python
#Optional apabila ingin memunculkan formula dari regresi linier
from scipy import stats

slope,intercept,r_value,p_value,std_err = stats.linregress(df_total['year'],df_total['total']) 
ax = sns.regplot(x='year',y='total',data=df_total,line_kws={'label':"y={0:.1f}x{1:.1f}".format(slope,intercept)}) #optional
ax.legend()
plt.show()
```

```python
# change color
ax = sns.regplot(x='year',y='total',data=df_total,color='blue')
ax.set_title('Regression Plot total Immigrants 1980-2010')
```

```python
#Change Color
ax = sns.regplot(x='year',y='total',data=df_total,color='blue',marker='+')
ax.set_title('Regreesion Plot Total Immigrants 1980-2010')
```

```python
#Change Size
fig = plt.figure(figsize=(15,10))
ax = sns.regplot(x='year',y='total',data=df_total,color='blue',marker='+')
ax.set_title('Regreesion Plot Total Immigrants 1980-2010')
plt.show()
```
