# Function berisi 3 parameter string yang akan dijoin menggunakan spasi dan convert menjadi reverse

```python
def str_joiner(a,b,c):
  a = a.lower()
  b = b.lower()
  c = c.lower()
  return (a+ ' ' +b+ ' ' +c)[::-1]
str_joiner('saya','mau','makan')
```

### A.n angka ganjil adalah = 7

### d. sum angka genap adalah = 100

```python
mylist = [31,23,31,40,50,67,74,88,91,10,11,13]

def list_summary(lst):
  nGenap = 0
  nGanjil = 0
  sumGenap = 0
  sumGanjil = 0
  for i in lst:
    if i % 2 == 0:
      nGenap +=1
      sumGenap +=i
    else:
      nGanjil +=1
      sumGanjil +=i
  print(nGanjil)
  print(nGenap)
  print(sumGanjil)
  print(sumGenap)

  print(list_summary(mylist))
```
