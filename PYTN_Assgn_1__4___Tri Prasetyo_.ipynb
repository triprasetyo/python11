# Area Plots

bar plot yang merupakan jenis grafik yang paling sering digunakan untuk merepresentasikan data numerik dan kategorik dalam bentuk bar. Lalu ada histogram yang digunakan untuk merepresentasikan distribusi frekuensi dan data numerik dengan batang

```python
london.head()
```

```python
london.plot(kind='area',stacked=False, figsize=(10,10))
plt.title('Top 5 Wilayah')
plt.ylabel('No. of Wilayah') #Judul sumbu y
plt.xlabel('Year') #Judul sumbu x
plt.show()
```

# Histogram

Histogram adalah jenis grafik yang menjelaskan frekuensi berdasarkan dua data numerik. Contohnya, data populasi berdasarkan umur dan jumlah proporsi laki-laki dan perempuan.

```python
london['year'].head()
```

```python
# plot dengan scripting layer
london['year'].plot(kind='hist')
plt.title('Histogram of wilayah from 195 in 2020')
plt.ylabel('Number of Wilayah(Frequency)')
plt.xlabel('Number of kategory utama')
plt.show()
```

# Pie Chart

Pie Chart merupakan grafik statistik yang berbentuk lingkaran. Lingkaran tersebut terbagi menjadi beberapa irisan dan luasnya tergantung pada proposi data. Diagram lingkaran digunakan untuk menunjukkan seberapa banyak dari setiap jenis kategori dalam dataset berbanding dengan keseluruhan

```python
london.head()
```

```python
# preparasi data piechart
london = london.groupby('wilayah').sum()
```

```python
london['Total'].plot(kind='pie',
                            figsize=(10,9),
                            autopct='%1.1f%%',
                            startangle=200,
                            shadow=True)
plt.title('Wilayah pada London')
plt.axis('Equal')
plt.show()
```

```python

colors_list = ['gold','yellowgreen','lightcoral','lightskyblue','lightgreen','pink']

london['year'].plot(kind='pie',
                          figsize=(15,6),
                          autopct='%1.1f%%',
                          startangle=90,
                          shadow=True,
                          labels=None,
                          pctdistance=1.12,
                          colors=colors_list)

plt.axis('equal')
plt.title('Wilayah Based on London pada Tahun 2013',y=1.12)
plt.show()
```

# Box Plot

Box plot adalah jenis visualisasi data yang secara statistik merepresentasikan distribusi data melalui lima dimensi utama, yaitu nilai minimum, kuartil 1, kuartil 2 (median), kuartil 3, dan nilai maksimum. Biasanya box plot digunakan untuk memeriksa keberadaan outlier dalam dataset.

```python
london.head()
```

```python
london.plot(kind='box',figsize=(8,6))
plt.title('Wilayah Tahun 1980-2013')
plt.ylabel('Number of Wilayah')
plt.show()
```

```python
london.describe()
```

# Subplots (Artist Layer)

Sublot merupakan suatu fungsi untuk membentuk banyak plot dalam figure.

```python
fig = plt.figure()
ax0 = fig.add_subplot(2,2,1) #tambah subplot 1 (1 row, 2 column, first plot)
ax1 = fig.add_subplot(1,2,2) #tambah subplot 2 (1 row , 2 columns, second plot)
london.plot(kind='box',color='blue',ax=ax0,figsize=(20,6),vert=False)
ax0.set_title('ax0')
ax0.set_ylabel('Countries')
ax0.set_xlabel('No. of Immigrants')

london.plot(kind='line',figsize=(20,6),ax=ax1)
ax1.set_title('ax1')
ax1.set_ylabel('Number of Immigrants')
ax1.set_xlabel('Years')
plt.suptitle('Visualisasi Data')

plt.show()
```
