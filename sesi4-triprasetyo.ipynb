# Buat function kalkulator dengan operator
### 'x' sebagai kali
### ':' sebagai bagi
### '+' sebagai tambah
### '-'sebagai kurang
### '^' sebagai pangkat

```python
c = '34'
c.isnumeric()
```

```python
try:
  int('sd')
except:
  print('bukan integer')
```

```python
def calc(a,b,c):
  if(a.isnumeric() == True) and (c.isnumeric() == True):
    a = int(a)
    b = int(c)
    if b == 'x':
      print('Hasil dari ',a,b,c, 'adalah ', a*c)
    elif b == '+':
      print('Hasil dari ',a,b,c, 'adalah ',a+c)
    elif b == ':':
      print('Hasil dari ',a,b,c, 'adalah ',a/c)
    elif b == '-':
      print('Hasil dari ',a,b,c, 'adalah ',a-c)
    elif b == '^':
      print('Hasil dari ',a,b,c, 'adalah ',a**c)
    else:
      print('operator', b, 'tidak ditemukan')
  else:
    print('salah')

a = (input('masukkan angka1 = '))
b = (input('masukkan operator = '))
c = (input('masukkan angka2 = '))

calc(a,b,c)
```

# Prime Number Generator Based on Interval

```python
# First, we will take the input:
lower_value = int(input("Please, Enter the lowest range value: "))
upper_value = int(input("Please, Enter the upper range value: "))

print("The prime numbers in the range are: ")
for number in range(lower_value,upper_value):
  if number > 1:
    for i in range(2, number):
      if(number% i) == 0:
        break
    else:
      print(number)
```

# Bilangan prima adalah bilangan yang hanya bisa dibagi 1 dan dirinya sendiri

```python
lower_limit = int(input('Lower Limit :'))
upper_limit = int(input("Upper Limit :"))
print(lower_limit, 'sampai',upper_limit)
for num in range(lower_limit,upper_limit+1):
    if num > 1:
        for i in range(2,num):#Filtering buat bilangan prima
            if (num%i)==0:
                break 
        else:
            print(num) #untuk kita dapat bilangan prima kondisi (num%i)==0 harus dilanggar
```
