## Conditional

### Python mendukung logical conditions yang biasa dipakai dalam matematika:

*   Equals:a == b
*   Not Equals: a != b
*   Less than: a < b
*   Less than or equal to: a <= b
*   Greather than: a > b
*   Greather than or equal to: a >= b
*   These conditions can be used in several ways, most commonly in "if statements" and loops





```python
x = 2
y = 3

if x < y:
  print("ya benar")
if x > y:
  print("ya benar")
if x == y:
  print("ya benar")
if x <= y:
  print('ya benar')
if 'we' in ['we','wa','wo']:
  print('ya benar')
```

# While Loop

```python
apel = 5
while apel > 0:
  apel-=1
  print(apel)
```

```python
apel = 5
apel -=1
print(4)
```

```python
apel = 5
while apel > 0:
  apel -=1
  if apel == 1:
    print('tinggal 1 hati hati ya')
  else:
    print(apel)
```

```python
# break digunakan untuk memberhentikan loop
apel = 5
while apel > 0:
  apel -=1
  if apel == 2:
    break
  print(apel)
print('loop ended')
```

```python
apel
```

```python
# continue digunakan untuk melanjutkan loop berdasarkan kondisi
apel = 5
while apel > 0:
  apel -=1
  if apel ==2:
    continue
  print(apel)
print('Loop ended')
```

```python
# else
apel = 10
while apel > 0:
  apel -=2
  print(apel)
else:
  print('loop ended')
```

```python
# break digunakan untuk memberhentikan loop
apel = 5
while apel > 0:
  apel -=1
  print(apel)
  if apel ==2:
    break
else:
    print('loop done')
```

# Nested If-Else & Loops

```python
# Nested If
umur = 42
gender = 'M'

if umur < 18:
  if gender == 'M':
    print('Son')
  else:
    print('Daugther')
if umur > 18 and umur < 65:
  if gender == 'M':
    print('Dad')
  else:
    print('Mom')
```

# For Loop

```python
a = [1,2,3,4,5,6]
for i in a:
  print(i)
```

```python
# For loop
b = ['Budi','Joko','Anwar']
for i in b:
  print(i)
```

```python
# Nested For Loop
b = ['Budi','Joko','Anwar']
for i in b:
  for j in i:
    print(j)
```

```python
# Nested For Loop
b = ['Budi','Joko','Anwar']
for i in b:
  if i == 'Joko':
    break
  for j in i:
    print(j)
```

# Menggunakan while loop buatlah program sederhana ketika bertemu dengan angka ganjil, maka print angkanya. ketika bertemu dengan angka genap, maka print 'genap' batas angka 1-20

```python
i = 1
while i <= 20:
  if i % 2==1:
    print(i)
  else:
    print('Genap')
  i+=1
```

```python
rows = int(input())
for i in range(1, rows + 1):
  for j in range(1, i + 1):
    print(j, end=' ')
  print('')
```
